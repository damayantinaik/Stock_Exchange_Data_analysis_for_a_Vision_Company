# 3. Calculate what the highest and lowest opening prices were for the stock in this period.

```python
opening_prices = [] 
for x in data2['datatable']['data']:
    opening_prices.append(x[2])
opening_prices[0:10]    
```

```python
max_opening_price = max(opening_prices)
print('The maximum opening price of the stock in the whole period is',  max_opening_price)
```

```python
min_opening_price = min(opening_prices)
print('The maximum opening price of the stock in the whole period is',  min_opening_price)
```

# 4. What was the largest change in any one day (based on High and Low price)?

```python
change_per_day_list =[]
for p in data2['datatable']['data']:
    change_per_day = max([p[2], p[3], p[4], p[5]])-min([p[2], p[3], p[4], p[5]])
    change_per_day_list.append( change_per_day)
print('The largest change in any one day(based on High and Low price): ', max(change_per_day_list))
```

# 5. What was the largest change between any two days (based on Closing Price)?

```python
closing_price_list = []
for i in data2['datatable']['data']:
    closing_price_list.append(i[5])

difference_between_two_days = []
for j in range(len(closing_price_list)-1):
    difference = closing_price_list[j+1] - closing_price_list[j]
    difference_between_two_days.append(difference)
    max_price_difference_between_two_days = max(difference_between_two_days)
print('Largest change between any two days: {0:.2f}'.format(max_price_difference_between_two_days))
```

# 6. What was the average daily trading volume during this year?

```python
from statistics import mean
volume_list = []
for element in data2['datatable']['data']:
    volume_list.append(element[6])
print(volume_list[:10])
avg_trading_volume = mean(volume_list)
#avg_trading_volume = float(sum(volume_list)/len(volume_list))
print('The average daily trading volume during the period: {0:.2f}'.format(avg_trading_volume))
```
